```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import plotly.graph_objs as go
```

```python
df_housing = pd.read_csv('C:\\Users\\iKerng\\Python\\School_DA\\ML_for_DA\\lesson1\\housing.csv')
df_housing
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_housing.drop('MEDV',axis=1)
                                                    , df_housing[['MEDV']]
                                                    , test_size=0.25
                                                    , random_state=42)

lin_reg_fit = LinearRegression().fit(X_train, y_train)
lin_reg_pred = lin_reg_fit.predict(X_test)

print('Среднеквадратичная ошибка (MSE) составляет ' + str(mean_squared_error(y_test, lin_reg_pred, squared=True)))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=np.arange(len(y_test)), y=y_test['MEDV'].to_list(), name='фактические значения'))
fig.add_trace(go.Scatter(x=np.arange(len(lin_reg_pred)), y=lin_reg_pred.T[0], name='предсказанные значения'))
fig.update_layout(legend_orientation="h")
fig.show()
```

```python
result = pd.DataFrame({'Feature': X_train.columns, 'FeatureImportance': lin_reg_fit.coef_[0]}).sort_values(
    by='FeatureImportance', ascending=False)
fig = go.Figure()
for i in range(len(result)):
    n = pd.DataFrame(result.iloc[i]).T
    fig.add_trace(go.Bar(x=n.FeatureImportance, y=n.Feature, orientation='h', name=n.Feature.iloc[0]))
fig.update_layout(legend_traceorder='reversed')
fig.show()
```
